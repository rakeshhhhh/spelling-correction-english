```python
from neuspell import CnnlstmChecker

# Initialize the checker
checker = CnnlstmChecker()

# Load the pretrained model (change the path if necessary)
checker.from_pretrained('cnn-lstm-probwordnoise')

# Function to correct spelling
def correct_spelling(sentence):
    corrected_sentence = checker.correct(sentence)
    return corrected_sentence

# User input sentence
user_input = input("Enter a sentence with potential spelling mistakes: ")

# Get the corrected sentence
corrected_sentence = correct_spelling(user_input)

# Display the result
print("Corrected Sentence:", corrected_sentence)
```

```python
from transformers import pipeline

# Initialize the spelling correction pipeline
fix_spelling = pipeline("text2text-generation", model="oliverguhr/spelling-correction-english-base")

# Function to correct spelling
def correct_spelling(user_sentence):
    clean_text = fix_spelling(user_sentence, max_length=2048)
    return clean_text[0]['generated_text']

# Take user input
user_input = input("Enter a sentence with potential spelling mistakes: ")

# Get the corrected sentence
corrected_sentence = correct_spelling(user_input)

# Display the result
print("Corrected Sentence:", corrected_sentence)
```
